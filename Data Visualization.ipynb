```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("DataBank.csv")

plt.scatter(data['job'], data['age'])

plt.title("Scatter Plot")

plt.xlabel('job')
plt.ylabel('age')

plt.show()
```
